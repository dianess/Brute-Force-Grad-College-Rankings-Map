```python
#Biological Sciences

executable_path = {"executable_path": r"C:\Users\Gabriel\Desktop\Classes_ASU\U_of_A\chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
url1 = "https://www.usnews.com/best-graduate-schools/top-science-schools/biological-sciences-rankings"
browser.visit(url1)

time.sleep(80)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

j = 0

cities = []
rankNumber = []
universityName = []


for i in soup.findAll('tr', attrs = {'class' : 'zebra TableTabular__TableRow-s9v6iyg-1 PNbVY'}):
    
    time.sleep(1.5)
    
    target1 = i.find('p', attrs={'class' : 'Paragraph-fqygwe-0 fJtpNK'})
    target2 = i.find('strong', class_='s144f3me-0-Strong-kDSDFS eULIZs')
    target3 = i.find('h3', class_='sc-bdVaJa kyuLHz')
    
    universityName.append(target3.text)
    cities.append(target1.text)
    rankNumber.append(target2.text)
    
    j = j + 1

    print (f'Processing record {j} of 192.')
    
    if j == 50:
        break
        
for i in soup.findAll('tr', attrs = {'class' : 'TableTabular__TableRow-s9v6iyg-1 PNbVY'}):
    
    time.sleep(1.5)
    
    target1 = i.find('p', attrs={'class' : 'Paragraph-fqygwe-0 fJtpNK'})
    target2 = i.find('strong', class_='s144f3me-0-Strong-kDSDFS eULIZs')
    target3 = i.find('h3', class_='sc-bdVaJa kyuLHz')
    
    universityName.append(target3.text)
    cities.append(target1.text)
    rankNumber.append(target2.text)
    
    j = j + 1

    print (f'Processing record {j} of 192.')
    
    if j == 50:
        break
        
print(cities)
print(rankNumber)
print(universityName)
```

```python
names = ('INSTNM', 'firstcity', 'Degree', 'Rank')
data = pd.DataFrame(columns = names)
data
```

```python
degrees = []

for i in range (20):
    degrees.append('Biological Sciences')

for i in range (20):
    degrees.append('Chemistry')
    
for i in range (20):
    degrees.append('Computer Sciences')

for i in range (20):
    degrees.append('Earth Sciences')
    
for i in range (20):
    degrees.append('Mathematics')

for i in range (20):
    degrees.append('Physics')
    
for i in range (20):
    degrees.append('Statistics')
    
degrees
```

```python
data['INSTNM'] = universityName
data['firstcity'] = cities
data['Degree'] = degrees
data['Rank'] = rankNumber

data.head()
```

```python
data['Rank'] = data['Rank'].apply(str).replace('#', '', regex=True)

data.head()
```

```python
data['CITY'] = data.firstcity.str.split(', ').str[0]
data['STATE'] = data.firstcity.str.split(', ').str[-1]

data
```
