```python
from bs4 import BeautifulSoup
from splinter.browser import Browser
import requests
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
#Law

executable_path = {"executable_path": r"C:\Users\Gabriel\Desktop\Classes_ASU\U_of_A\chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
url1 = "https://www.usnews.com/best-graduate-schools/top-law-schools/law-rankings/"
browser.visit(url1)

time.sleep(80)

cities = []
rankNumber = []
universityName = []

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

j = 0


for i in soup.findAll('div', attrs = {'class' : 's85n6m5-0-Box-cwadsP dlYdgX'}):
    
    time.sleep(1.5)
    
    target1 = i.find('p', attrs={'class' : 'Paragraph-fqygwe-0 bstttc'})
    target2 = i.find('div', class_='ranked RankList__Rank-qghlf4-2 hSEzjh')
    target3 = i.find('a', class_='Anchor-s1mkgztv-0 gPdGYK')
    
    universityName.append(target3.text)
    cities.append(target1.text)
    rankNumber.append(target2.text)
    
    j = j + 1

    print (f'Processing record {j} of 192.')
    
    if j == 50:
        break
        
print(cities)
print(rankNumber)
print(universityName)
```

```python
names = ('INSTNM', 'firstcity', 'Degree', 'Rank')
data = pd.DataFrame(columns = names)
data
```

```python
degrees = []

for i in range (20):
    degrees.append('Law')

for i in range (20):
    degrees.append('Business')
    
for i in range (20):
    degrees.append('Education')

for i in range (20):
    degrees.append('Engineering')
    
for i in range (20):
    degrees.append('Medicine')

for i in range (20):
    degrees.append('Nursing')
    
for i in range (20):
    degrees.append('Fine Arts')

for i in range (20):
    degrees.append('Library and Information Studies')
    
for i in range (20):
    degrees.append('Public Affairs')
    
degrees
```

```python
data['INSTNM'] = universityName
data['firstcity'] = cities
data['Degree'] = degrees
data['Rank'] = rankNumber

data.head()
```

```python
data['Rank'] = data['Rank'].apply(str).replace('#', '', regex=True)

data.head()
```

```python
data['CITY'] = data.firstcity.str.split(', ').str[0]
data['STATE'] = data.firstcity.str.split(', ').str[-1]

data
```
