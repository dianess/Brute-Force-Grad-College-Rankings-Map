# Graduate Rankings Map


## Convert ranking data to html

```python
# Dependencies
import csv
import pandas as pd
```

```python
# Read the CSV file and create a dataframe
all_rankings_df= pd.read_csv('allrankingsfordisplay.csv')
all_rankings_df.head()
```

```python
# Rename columns
all_rankings_renamed=all_rankings_df.rename(columns={"INSTNM":"Institution Name", "CITY":"City", "STATE":"State", "Degree":"Degree", "Rank":"Rank", "LAT": "Latitude", "LON": "Longitude"})
all_rankings_renamed.head()
```

```python
# Remove the unnecessary index
all_rankings=all_rankings_renamed.set_index('Institution Name')
all_rankings.head()
```

```python
# Convert the pandas dataframe to html and export as "dow_jones_html_data.html"
all_rankings.to_html('all_rankings.html')
```
